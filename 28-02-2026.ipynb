```python
### find and remove the brackets

start = df.loc[1, "Name"].find("(")
end = df.loc[1, "Name"].find(")")
```

```python
df.loc[1, "Name"][0:start] + df.loc[1, "Name"][end+1:]
```

```python
def remove_brackets(name):
    start = name.find("(")
    end = name.find(")")
    new_name = name[0:start] + name[end+1:]
    new_name = new_name.strip()
    return new_name
    
```

```python
df["Name_trimmed"] = df["Name"].apply(remove_brackets)
df["Name_trimmed"].head()
```

```python
df.loc[1, "Name_trimmed"]
```

```python
#### replace Substring with Regex : regular expression

# *  means zero or more
# . match any character
# # r"\(.*\)"
# windows : \\

df["Name"].str.replace(r"\\(.*\\)", "Hello", regex=False).head()
```

## Categorical Data in Pandas

```python
class_data = {
"Name" : ["A","B","C","D"],
"Groups":["G1","G2","G3","G4"]

}
```

```python
class_df = pd.DataFrame(class_data)
class_df



```

```python
class_df.info()
```

```python
class_df["Group_cat"] = pd.Categorical(
    class_df["Groups"],
    categories=["G3", "G1", "G2", "G4"], ## CUSTOM ORDER DONE LOWER TO HIGHER ascending order
    #["G3"= 1, "G1"=2, "G2"=3, "G4"= 4]
    ordered=True
)
```

```python
class_df.info()
```

```python
class_df["Group_cat"].cat.categories
```

```python
class_df["Group_cat"] >= "G4"
#["G3"= 1, "G1"=2, "G2"=3, "G4"= 4]
# if g3 greater than g4 =false
# if g1 greater than g4 =false
# if g2 greater than g4 =false
# if g4 greater than g4 = true  based on the ordering
```

```python
titanic_dataset_github_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_dataset_github_url)

df.head()
```

## Date functionality in pandas

```python
import pandas as pd

# Creating a dictionary with your data
data = {
    'event_id': [101, 102, 103, 104, 105],
    'event_name': ['Product Launch', 'Quarterly Review', 'Team Building', 'Client Pitch', 'System Maintenance'],
    'date': ['2026-03-01', '2026-03-15', '2026-04-05', '2026-04-20', '2026-05-10'],
    'priority': ['High', 'Medium', 'Low', 'High', 'Critical']
}

```

```python
# Initializing the DataFrame
df = pd.DataFrame(data)
df


```

```python
# Crucial step: Convert the 'date' column to datetime objects
df['date_pandas'] = pd.to_datetime(df['date'])
df
```
